```python
td = td[['Overall Qual', 'Gr Liv Area', 'Garage Cars', 'Garage Area',
                     'Total Bsmt SF', '1st Flr SF', 'Year Built', 'Year Remod/Add',
                     'Full Bath', 'TotRms AbvGrd', 'Neighborhood', 'SalePrice']]
```

```python
td.shape
```

```python
td.isnull().sum()
```

```python
td.dropna(inplace=True)
```

```python
td.isnull().sum()
```

```python
import datetime
current_year = datetime.datetime.now().year
td['Age of House'] = current_year - td['Year Remod/Add']
td.loc[td['Year Remod/Add'] == 0, 'Age of House'] = current_year - td.loc[td['Year Remod/Add'] == 0, 'Year Built']
```

```python
td[['Year Built', 'Year Remod/Add', 'Age of House']].head()
```

```python
processed_dataset_path = "../Data_Collection/processed_ames_housing.csv"
td.to_csv(processed_dataset_path, index=False)
```
